**Task**

- Step-1: take prevailing wage column: **visa_df['prevailing_wage']**

- step-2: calculate the min value of prevailing wage column: **visa_df['prevailing_wage'].min**

- step-3: calculate the max value of prevailing wage column: **visa_df['prevailing_wage'].max**

- step-4: calculate the numerator : step1-step2 :Nr= **visa_df['prevailing_wage']-min**

- step-5: Dr= step 3 - step 2

- step-6: divide the  step4/step5

**package name**: **MinMaxScalar**

```python
visa_df=pd.read_csv(r"C:\Users\gunda\Documents\Documents\Naresh IT\Data Science\Data files\visadataset.csv")
visa_df.head(2)
```

```python
x_min=visa_df['prevailing_wage'].min()
x_max=visa_df['prevailing_wage'].max()
Nr= visa_df['prevailing_wage']-x_min
visa_df['prevailing_wage_min_max']= Nr/(x_max-x_min)
visa_df[['prevailing_wage_min_max','prevailing_wage']]
```

```python
minn=visa_df['prevailing_wage'].min()
maxx=visa_df['prevailing_wage'].max()
Nr=visa_df['prevailing_wage']-minn
Dr=maxx-minn
visa_df['prevailing_wage_min_max']=Nr/Dr
visa_df[['prevailing_wage_min_max','prevailing_wage']]
```
