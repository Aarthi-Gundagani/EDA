**Create data frame using dictionary**

```python
import pandas as pd
names=['Avinash','Aakash','Aadhya']
age=[20,25,30]
pd.DataFrame(zip(names,age),columns=['Names','Age'])
```

```python
dict1={'Names':['Avinash','Aakash','Aadhya'],
       'Ages':[25,30,35]}
dict1
```

```python
pd.DataFrame(dict1)
# keys becomes columns
# values becomes rows
# No need to provide columns names separately
```

```python
dict2={'Names': 'Avinash',
       'Ages':25,
       'City':'Hyd'}
pd.DataFrame(dict2)
```

```python
dict2={'Names': 'Avinash',
       'Ages':25,
       'City':'Hyd'}
pd.DataFrame(dict2,index=[1,2,3])
# we must pass the index, when we dont use list. scalar values means single value
```
