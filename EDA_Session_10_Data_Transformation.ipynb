**Data Transformations**

- Generally, data has 3 types of distributions

   - Normal distribution
 
   - Positive skewed
 
   - Negative skewed
 
- All the maths concepts developed by assumption as Data follows Normal Distributrion

- Data skew happens because of many reasons, in that one reason is outliers

- once we perform the outlier analysis, then we can apply transformation methods, if the data does not follows normality

- We have some transformation methods mentioned below

   - Log transformation
 
   - Exponential transformation
 
   - Reciprocal transformation
 
   - Square root transformation
 
   - Power trransformation

**Log transformation**

- Log transformation means performing the logarithm operations on original data

- It is an approach to convert into normal distribution

- Log means natural logarithm, which means base=e

```python
x=2
np.log(x)
```

```python
log_data=np.log(exp_data)
log_data
```

```python
plt.hist(log_data,bins=20,label='Logarithm')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(exp_data,bins=30,label='Expoenential')
plt.legend()
plt.subplot(1,2,2).hist(log_data,bins=30,label='Logarithm')
plt.legend()
plt.show()
```

**Reciprocal Transformation**

- Reciprocal will fail when values are low

- For example, x=0 then 1/0 will fail

```python
x=2
1/x
```

```python
reciprocal_data=np.reciprocal(exp_data)
plt.hist(reciprocal_data,bins=20,label='Reciprocal')
plt.show()
```

**Squareroot transformation**

```python
sqrt_data=np.sqrt(exp_data)
plt.hist(sqrt_data,bins=20,label='Square Root Data')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.subplot(1,2,1).hist(log_data,bins=20,label='Logarithm')
plt.legend()
plt.subplot(1,2,2).hist(sqrt_data,bins=30,label='Square Root Data')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,4,1).hist(log_data,bins=20,label='Logarithm')
plt.legend()
plt.subplot(1,4,2).hist(exp_data,bins=20,label='Exponential')
plt.legend()
plt.subplot(1,4,3).hist(reciprocal_data,bins=20,label='Reciprocal Data')
plt.legend()
plt.subplot(1,4,4).hist(sqrt_data,bins=30,label='Square Root Data')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer
plt.figure(figsize=(12,5))
power = PowerTransformer(method='yeo-johnson')
exp_data1=exp_data.reshape(-1,1)
data_trans=power.fit_transform(exp_data1)
plt.subplot(1,2,1).hist(data_trans,bins=30,label='Yeo-Johnson')
plt.legend()

power= PowerTransformer(method='box-cox')
exp_data2=exp_data.reshape(-1,1)
data_trans=power.fit_transform(exp_data2)
plt.subplot(1,2,2).hist(data_trans,bins=30,label='Box-Cox')
plt.legend()
plt.show()
```
